```
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt


#Tasks: To calculate the t-statistic and p-value of Data = [5.2, 5.4, 5.1, 5.6, 5.3, 5.8] and Hypothesized mean = 5.5. Interpret and visualize the result at a 0.05 significance level.

data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
hypothesized_mean = 5.5

t_statistic, p_value = stats.ttest_1samp(data, hypothesized_mean)

print(f'Sample Mean: {np.mean(data):.2f}')
print(f'Hypothesized Mean: {hypothesized_mean:.2f}')
print(f't-Statistic: {t_statistic:.2f}')
print(f'Degrees of Freedom: {len(data) - 1}')
print(f'p-Value: {p_value:.4f}')

alpha = 0.05
if p_value < alpha:
    print(f'Reject the null hypothesis. The average is significantly different from {hypothesized_mean:.2f}.')
else:
    print(f'Fail to reject the null hypothesis. The average is not significantly different from {hypothesized_mean:.2f}.')

plt.figure()
plt.bar(1, np.mean(data), color="green", label='Sample Mean')
plt.errorbar(1, np.mean(data), yerr=np.std(data), fmt='k', linewidth=1.5)
plt.axhline(y=hypothesized_mean, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title('Sample Mean vs Hypothesized Mean')
plt.xlim(0.5, 1.5)
plt.xlabel('Sample')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
